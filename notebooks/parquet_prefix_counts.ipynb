# Contagem de parquets por prefixo de 6 letras

Lista todos os arquivos `.parquet` em `data/experiments/single` e conta quantos existem por prefixo de 6 caracteres do nome do arquivo.

```python
from collections import Counter
from pathlib import Path

REPO_ROOT = Path.cwd().parent
DATA_DIR = REPO_ROOT / "data" / "experiments" / "single"

parquet_files = sorted(p.name for p in DATA_DIR.glob("*.parquet"))
print(f"Total de arquivos .parquet: {len(parquet_files)}")
parquet_files[:10]
```

```python
counts = Counter(name[:6] for name in parquet_files)

for prefix, n in sorted(counts.items(), key=lambda kv: (-kv[1], kv[0])):
    print(f"{prefix}  {n}")
```

```python
import pandas as pd

df = (
    pd.DataFrame(counts.items(), columns=["prefix", "count"])
    .sort_values(["count", "prefix"], ascending=[False, True])
    .reset_index(drop=True)
)
df
```
