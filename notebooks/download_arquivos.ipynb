```python
import os
from pathlib import Path
from google.cloud import storage
from tqdm.auto import tqdm

# Configurações
MEU_BUCKET = "ua-sa-nsga2"
PREFIXO = "dataframes/"
DIRETORIO_BASE_LOCAL = "data"

def baixar_arquivos_gcs_com_data(nome_bucket, prefixo, base_local):
    """
    Lista e baixa arquivos de um bucket GCS preservando a estrutura de diretórios,
    além de exibir a data de modificação (mod_date) de cada arquivo em UTC.
    """
    # Instancia o cliente (autenticação feita pela Service Account da VM)
    client = storage.Client()
    bucket = client.bucket(nome_bucket)
    
    # Lista todos os blobs com o prefixo
    blobs = list(bucket.list_blobs(prefix=prefixo))
    print(f"Foram encontrados {len(blobs)} arquivos no bucket.\n")
    
    # Executa o loop de download com barra de progresso
    for blob in tqdm(blobs, desc="Baixando dataframes"):
        # Ignora "diretórios virtuais" caso existam no GCS
        if blob.name.endswith('/'):
            continue
            
        # Pega a data de atualização e formata
        data_modificacao = blob.updated
        data_formatada = data_modificacao.strftime("%d/%m/%Y às %H:%M:%S (UTC)")
        
        # Exibe o log do arquivo atual
        print(f"Lendo: {blob.name} | Modificado em: {data_formatada}")
        
        # Constrói o caminho local (ex: data/dataframes/nome_do_problema/arquivo.parquet)
        caminho_local = Path(base_local) / blob.name
        
        # Cria a estrutura de pastas localmente (equivalente a mkdir -p)
        caminho_local.parent.mkdir(parents=True, exist_ok=True)
        
        # Executa o download salvando no disco da VM
        blob.download_to_filename(str(caminho_local))
        
    print(f"\nDownload concluído! Os dados foram salvos no diretório '{base_local}'.")

# Executa a função
baixar_arquivos_gcs_com_data(MEU_BUCKET, PREFIXO, DIRETORIO_BASE_LOCAL)
```
