# Limpeza: deletar parquets com sufixo `_clean`

Apaga todos os arquivos `*_clean.parquet` em `data/experiments/single/`.

Esses arquivos foram criados por uma convenção antiga de nomeação (`noisy_problem=False` → sufixo `_clean`). A convenção atual é o oposto:
- `noisy_problem=False` (default) → **sem sufixo**
- `noisy_problem=True` → sufixo `_noisy`

O notebook tem 3 células: **listar**, **confirmar**, **deletar**. Rode-as em ordem.

```python
# 1) Listar os arquivos *_clean.parquet em data/experiments/single/
from pathlib import Path

# Resolve a pasta a partir da raiz do repo (notebook fica em notebooks/)
REPO_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
TARGET_DIR = REPO_ROOT / 'data' / 'experiments' / 'single'

print(f'Procurando em: {TARGET_DIR}')
print(f'Existe? {TARGET_DIR.exists()}')

to_delete = sorted(TARGET_DIR.glob('*_clean.parquet'))
print(f'\nEncontrados {len(to_delete)} arquivos *_clean.parquet:')
for p in to_delete:
    print(f'  - {p.name}  ({p.stat().st_size / 1024:.1f} KB)')
```

```python
# 2) Conferir total (em KB e contagem) antes de apagar
total_bytes = sum(p.stat().st_size for p in to_delete)
print(f'Total a deletar: {len(to_delete)} arquivo(s)  /  {total_bytes / 1024:.1f} KB')
print('\nSe a lista acima estiver correta, rode a proxima celula para deletar.')
```

```python
# 3) Deletar (irreversivel)
deleted, failed = [], []
for p in to_delete:
    try:
        p.unlink()
        deleted.append(p.name)
    except Exception as e:
        failed.append((p.name, str(e)))

print(f'Deletados: {len(deleted)}')
for name in deleted:
    print(f'  ✓ {name}')

if failed:
    print(f'\nFalhas: {len(failed)}')
    for name, err in failed:
        print(f'  ✗ {name}: {err}')

# Confirma o estado final
remaining = sorted(TARGET_DIR.glob('*_clean.parquet'))
print(f'\nRestantes com sufixo _clean: {len(remaining)}')
print(f'Total de arquivos em {TARGET_DIR.name}/: {len(list(TARGET_DIR.glob("*.parquet")))}')
```
